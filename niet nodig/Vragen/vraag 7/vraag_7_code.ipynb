```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import csv
from collections import Counter

df1 = pd.read_csv("DataGun.csv")

states = [
    'Alabama','Alaska','Arizona','Arkansas','California','Colorado',
    'Connecticut','Delaware','Florida','Georgia','Hawaii','Idaho', 
    'Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana',
    'Maine','Maryland','Massachusetts','Michigan','Minnesota',
    'Mississippi', 'Missouri','Montana','Nebraska','Nevada',
    'New Hampshire','New Jersey','New Mexico','New York',
    'North Carolina','North Dakota','Ohio',    
    'Oklahoma','Oregon','Pennsylvania','Rhode Island',
    'South Dakota','Tennessee','Texas','Utah',
    'Vermont','Virginia','Washington','West Virginia',
    'Wisconsin','Wyoming'
]

df2 = df1.loc[df1['gun_type'].str.contains("NaN")==False]
df3 = df2.loc[df2['gun_type'].str.contains("Unknown")==False]
df4 = df3.loc[df3['gun_type'].str.contains("Other")==False]

wapens = []
for i in df4['gun_type']:
    woord = i.split("::")
    for j in woord:
        if len(j) > 2 and 'Unknown' not in j:
            if "||" in j[len(j)-3:]:
                x = j[:len(j)-3]
                if x != "Unknown":
                    if x not in wapens and "|" not in x and ":" not in x:
                        wapens.append(x)
            elif "||" in j[len(j)-4:]:
                y = j[:len(j)-4]
                if y != "Unknown":
                    if y not in wapens and "|" not in  y and ":" not in y:
                        wapens.append(y)
            elif "||" in j[len(j)-5:]:
                z = j[:len(j)-5]
                if z != "Unknown":
                    if z not in wapens and "|" not in z and ":" not in z:
                        wapens.append(z)
            else:
                if j != "Unknown":
                    if j not in wapens and "|" not in j and ":" not in j:
                        wapens.append(j)
            
print(wapens, '\n')

for i in states:
    print(i)
    df = df4.loc[df4['state'].str.contains(i)]
    a = pd.Series(Counter([y for x in df['gun_type'].str.findall('(?<=\:\:)\w+\s*\-*\.*\w+\s*\w*(?=||)') for y in x]))
    b = a.drop(["Handgun","Rifle","Shotgun","9mm"])
    c = b.sort_values(ascending=False)
    print(c.keys()[0])
    print(c.keys()[1])
    print(c.keys()[2])
    
```
