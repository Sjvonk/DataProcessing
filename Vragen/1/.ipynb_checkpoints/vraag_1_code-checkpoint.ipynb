```python
# import matplotlib.pyplot as plt
# %matplotlib inline
# # Load library
import numpy as np
import pandas as pd
import csv
import plotly.plotly as py
import plotly.graph_objs as go
import plotly

# Print map van de VS met number_killed, number_injured, total_casualties en kills_per_citizen

plotly.tools.set_credentials_file(username='MCampman', api_key='enJkmcytGcgSw3Tk37vO')

df = pd.read_csv('vraag1.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(255,255,255)'],[0.2, 'rgb(255,204,204)'],[0.4, 'rgb(255,153,153)'],\
            [0.6, 'rgb(255,102,102)'],[0.8, 'rgb(255,51,51)'],[1.0, 'rgb(255,0,0)']]


num_cit = pd.to_numeric(df['citizens_per_state'])
num_kill = pd.to_numeric(df['number_killed'])
kill_per_cit = num_kill/num_cit
kill_per_cit1 = kill_per_cit.astype(str)

df['text'] = df['state'] + '<br>' +\
    'Number killed: '+df['number_killed'] + '<br>' +\
    'Number injured: ' +df['number_injured'] + '<br>' +\
    'Total casualties: ' +df['total_casualties'] + '<br>' +\
    'Kills per citizen: '+ kill_per_cit1

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['code'],
        z = df['total_casualties'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Number killed")
        ) ]

layout = dict(
        title = '2013-2018 differences between states',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )


# Bereken number_killed en number_injured per jaar

# df1 = pd.read_csv('DataGun.csv')
# df2 = df1.set_index("date", drop=False)

# df13 = df2.loc["1/1/2013":"12/31/2013", "state":"state_senate_district"]
# df14 = df2.loc["1/1/2014":"12/31/2014", "state":"state_senate_district"]
# df15 = df2.loc["1/1/2015":"12/31/2015", "state":"state_senate_district"]
# df16 = df2.loc["1/1/2016":"12/31/2016", "state":"state_senate_district"]
# df17 = df2.loc["1/1/2017":"12/31/2017", "state":"state_senate_district"]
# df18 = df2.loc["1/1/2018":"3/31/2018", "state":"state_senate_district"]
        

# number_killed_2013 = df13["n_killed"].sum()
# number_injured_2013 = df13["n_injured"].sum()

# number_killed_2014 = df14["n_killed"].sum()
# number_injured_2014 = df14["n_injured"].sum()

# number_killed_2015 = df15["n_killed"].sum()
# number_injured_2015 = df15["n_injured"].sum()

# number_killed_2016 = df16["n_killed"].sum()
# number_injured_2016 = df16["n_injured"].sum()

# number_killed_2017 = df17["n_killed"].sum()
# number_injured_2017 = df17["n_injured"].sum()

# number_killed_2018 = df18["n_killed"].sum()
# number_injured_2018 = df18["n_injured"].sum()

# print('n_killed 2013:', number_killed_2013, '\n')
# print('n_injured 2013:', number_injured_2013, '\n')

# print('n_killed 2014:', number_killed_2014, '\n')
# print('n_injured 2014:', number_injured_2014, '\n')

# print('n_killed 2015:', number_killed_2015, '\n')
# print('n_injured 2015:', number_injured_2015, '\n')

# print('n_killed 2016:', number_killed_2016, '\n')
# print('n_injured 2016:', number_injured_2016, '\n')

# print('n_killed 2017:', number_killed_2017, '\n')
# print('n_injured 2017:', number_injured_2017, '\n')

# print('n_killed 2018:', number_killed_2018, '\n')
# print('n_injured 2018:', number_injured_2018, '\n')



# Voeg number_killed, number_injured en casualties toe aan een nieuwe csv file

# df111 = pd.read_csv('file_name.csv') 
# df222 = pd.read_csv('vraag1.csv')

# df111.insert(value = [kill for kill in df111['n_killed']], loc= df111.shape[1], column = 'number_killed')
# df111.insert(value = [inj for inj in df111['n_injured']], loc= df111.shape[1], column = 'number_injured')
# df111.insert(value = [(kill + inj) for kill, inj in zip(df111['n_killed'], 
#           df111['n_injured'])], loc= df111.shape[1], column = 'casualties')

# df222['number_killed'] = df111.groupby(['state'])['number_killed'].transform(sum)
# df222['number_injured'] = df111.groupby(['state'])['number_injured'].transform(sum)
# df222['total_casualties'] = df111.groupby(['state'])['casualties'].transform(sum)

# df222.to_csv('vraag1.csv')


```
